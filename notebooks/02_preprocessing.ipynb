# 02 – Data Preprocessing

Min-Max normalization, rolling features (window=60) and 70/15/15 split.

```python
import sys, os; sys.path.insert(0, os.path.abspath('..'))
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from src.utils.preprocessing import generate_synthetic_dataset, preprocess_pipeline, FEATURE_COLUMNS
df = generate_synthetic_dataset(n_samples=120_000, seed=42)
feature_cols = [c for c in FEATURE_COLUMNS if c in df.columns]
train, val, test, scaler, all_cols = preprocess_pipeline(df, feature_cols)
print(f'Train: {len(train)}, Val: {len(val)}, Test: {len(test)}')
print(f'Total features after rolling: {len(all_cols)}')
```

```python
print('Min after norm:', train[feature_cols].min().min())
print('Max after norm:', train[feature_cols].max().max())
```

```python
col = feature_cols[0]
fig, axes = plt.subplots(3, 1, figsize=(14, 8), sharex=True)
axes[0].plot(train[col].values[:2000], lw=0.8); axes[0].set_title(f'{col} normalized')
axes[1].plot(train[f'{col}_roll_mean'].values[:2000], lw=0.8, color='orange'); axes[1].set_title('Rolling mean')
axes[2].plot(train[f'{col}_zscore'].values[:2000], lw=0.8, color='green'); axes[2].set_title('Z-score')
plt.tight_layout(); plt.savefig('../results/preprocessing_rolling_features.png', dpi=100); plt.show()
```
